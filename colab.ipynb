## Explore dataset

```
%cd /gdrive/Team\ Drives/umayux/Research/NLP/humor_detection/
import pandas as pd
from collections import Counter

data = pd.read_csv('./data/haha_2019_train.csv', sep=',')
data.head()

fo = open('./data/data.tsv', 'w')
for ix,i in data.iterrows():
    
    if i['is_humor'] == 0:
        target = 'no_humor'
    else:
        target='humor'
        
    fo.write('{}\t{}\n'.format(i['text'].strip().replace('\n', ' ').replace('\r', ' '), target))

fo.close()
```
